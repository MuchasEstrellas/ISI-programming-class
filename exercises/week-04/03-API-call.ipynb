# Python API Call


```python
import requests
import json
import pprint

url = "https://api.coinone.co.kr/ticker/"
response = requests.get(url)
data = json.loads(response.content)
pprint.pprint(data)
```

```python
currencies = ['btc', 'bch', 'eth', 'etc', 'xrp', 'qtum', 'iota', 'ltc', 'btg', 
              'omg', 'eos', 'data', 'zil', 'knc', 'zrx', 'xtz', 'all']
for currency in currencies:
    response = requests.get(url+"?currency="+currency)
    data = json.loads(response.content)
    pprint.pprint(data)
```
