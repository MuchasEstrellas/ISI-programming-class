# Matplotlib

```python
import matplotlib.pyplot as plt

#when you use jupyter notebook.
%matplotlib inline
```

## 1. Basic

```python
import numpy as np
x = np.linspace(1, 10, 50) #n
a = np.log(x) #logn
b = x ** 2 #n^2
c = np.log(x)*x #nlogn 
d = 2**x #2^n
```

```python
plt.plot(x, a, 'r') # 'r' is the color red
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Title')
plt.show()
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,3,1)
plt.plot(x, b, 'r*-');
plt.subplot(1,3,2)
plt.plot(x, c, 'b--');
plt.subplot(1,3,3)
plt.plot(x, d, 'g+-');

#What's this?
plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, a, label="logN")
ax.plot(x, b, label="N^2")
ax.plot(x, c, label="NlogN")
ax.plot(x, d, label="2^N")
ax.legend()
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.4,0.3,0.3])

ax1.plot(x, a, label="logN")
ax1.plot(x, b, label="N^2")
ax1.legend()

ax2.plot(x, c, label="NlogN")
ax2.plot(x, d, label="2^N")
ax2.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, a, label="logN")
ax.plot(x, b, label="N^2")
ax.plot(x, c, label="NlogN")
ax.plot(x, d, label="2^N")
ax.set_ylim([0,100])
ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.set_title("Title")
ax.plot(x, a, label="logN")
ax.plot(x, b, label="N^2")
ax.plot(x, c, label="NlogN")
ax.plot(x, d, label="2^N")
ax.set_ylim([0,100])
ax.set_xlim([3,7])
ax.legend()
```

### 2. Other plots

Code can be found from [here](https://matplotlib.org/gallery/index.html)

```python
plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = ['Tom', 'Dick', 'Harry', 'Slim', 'Jim']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center',
        color='green', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
```

```python
fig, ax = plt.subplots()
for color in ['red', 'green', 'blue']:
    n = 750
    x, y = np.random.rand(2, n)
    scale = 200.0 * np.random.rand(n)
    ax.scatter(x, y, c=color, s=scale, label=color,
               alpha=0.3, edgecolors='none')

ax.legend()
ax.grid(True)

plt.show()
```

```python
N_points = 100000
n_bins = 20

# Generate a normal distribution, center at x=0 and y=5
x = np.random.randn(N_points)
y = .4 * x + np.random.randn(100000) + 5

fig, axs = plt.subplots(1, 2, sharey=True)

# We can set the number of bins with the `bins` kwarg
axs[0].hist(x, bins=n_bins)
axs[1].hist(y, bins=n_bins)
```
